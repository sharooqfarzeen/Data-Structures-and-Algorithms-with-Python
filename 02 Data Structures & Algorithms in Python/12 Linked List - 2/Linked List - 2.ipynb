## Reverse LL - Recursive

Given a singly linked list of integers, reverse it using recursion and return the head to the modified list. You have to do this in O(N) time complexity where N is the size of the linked list.

 Note :
No need to print the list, it has already been taken care. Only return the new head to the list.

```python
from sys import stdin, setrecursionlimit
setrecursionlimit(10 ** 6)

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None



def reverseLinkedListRec(head, previous, current, length):
  if length < 2:
    return head
  if current == length-1:
    head.next = previous
    return head
  newHead = reverseLinkedListRec(head.next, head, current+1, length)
  head.next = previous
  return newHead

#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))
    length = len(datas)-1
    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head, length




def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head, length = takeInput()

    newHead = reverseLinkedListRec(head, None, 0, length)
    printLinkedList(newHead)

    t -= 1
```

## Merge 2 sorted Linked Lists

You have been given two sorted(in ascending order) singly linked lists of integers.

Write a function to merge them in such a way that the resulting singly linked list is also sorted(in ascending order) and return the new head to the list.

Note :
Try solving this in O(1) auxiliary space.

No need to print the list, it has already been taken care.

```python
from sys import stdin

class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None

def insert(data, before, after, position):
  if position == 0:
    data.next = after
  else:
    before.next = data
    data.next = after

def mergeTwoSortedLinkedLists(head1, head2):
  if head1 is None:
    if head2 is None:
      return head1
    else:
      return head2
  elif head2 is None:
    return head1

  curr1 = head1
  prev1 = None
  curr2 = head2
  prev2 = None
  position = 0
  if curr2.data <= curr1.data:
    mainHead = head2
  else:
    mainHead = head1

  while (curr1 is not None) and (curr2 is not None):
    if curr1.data < curr2.data:
      prev1 = curr1
      curr1 = curr1.next
    else:
      next2 = curr2.next
      insert(curr2, prev1, curr1, position)
      prev1 = curr2
      curr2 = next2
    position += 1
  if curr1 is None:
    prev1.next = curr2
  return mainHead



#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head




def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


# Main
t = int(stdin.readline().rstrip())

while t > 0 :

    head1 = takeInput()
    head2 = takeInput()

    newHead = mergeTwoSortedLinkedLists(head1, head2)
    printLinkedList(newHead)

    t -= 1
```

```python
from sys import stdin, setrecursionlimit
setrecursionlimit(10 ** 6)

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None

def mergeSort(head):
    # Base case: if the linked list has zero or one element, it is already sorted
    if head is None or head.next is None:
        return head

    # Find the middle of the linked list
    mid = findMiddle(head)
    midNext = mid.next
    mid.next = None  # Split the list into two halves

    # Recursively sort the two halves
    left = mergeSort(head)
    right = mergeSort(midNext)

    # Merge the sorted halves
    sortedList = merge(left, right)

    return sortedList

def findMiddle(head):
    slow = head
    fast = head.next

    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next

    return slow

def merge(left, right):
    if left is None:
        return right
    if right is None:
        return left

    if left.data < right.data:
        result = left
        result.next = merge(left.next, right)
    else:
        result = right
        result.next = merge(left, right.next)

    return result

#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head


def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()

    newHead = mergeSort(head)
    printLinkedList(newHead)

    t -= 1
```

## Even After Odd - Linked List

For a given singly linked list of integers, arrange the nodes such that all the even number nodes are placed after the all odd number nodes. The relative order of the odd and even terms should remain unchanged.

Note :
1. No need to print the linked list, it has already been taken care. Only return the new head to the list.
2. Don't create a new linked list.
3.  Just change the data, instead rearrange the provided list.

```python
from sys import stdin

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None


def evenAfterOdd(head):
    if head is None:
        return None

    # Separate odd and even nodes
    odd_head, even_head = None, None
    odd_tail, even_tail = None, None

    current = head
    while current is not None:
        if current.data % 2 == 0:  # Even node
            if even_head is None:
                even_head = current
                even_tail = current
            else:
                even_tail.next = current
                even_tail = current
        else:  # Odd node
            if odd_head is None:
                odd_head = current
                odd_tail = current
            else:
                odd_tail.next = current
                odd_tail = current
        current = current.next

    # If no odd nodes, return even nodes
    if odd_head is None:
        return even_head

    # If no even nodes, return odd nodes
    if even_head is None:
        return odd_head

    # Append even nodes after odd nodes
    odd_tail.next = even_head
    even_tail.next = None

    return odd_head

#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head


#to print the linked list
def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()
    newHead = evenAfterOdd(head)
    printLinkedList(newHead)

    t -= 1
```

## Delete Every N Nodes

You have been given a singly linked list of integers along with two integers, 'M,' and 'N.' Traverse the linked list such that you retain the 'M' nodes, then delete the next 'N' nodes. Continue the same until the end of the linked list. Indexing starts from 1.

To put it in other words, in the given linked list, you need to delete N nodes after every M nodes.

Note :
No need to print the list, it has already been taken care. Only return the new head to the list. You can return null in case where all nodes will be deleted.

```python
from sys import stdin

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None



def skipMdeleteN(head, M, N):
    if head is None or M == 0:
        return None

    current = head
    while current:
        # Skip M nodes
        for _ in range(M - 1):
            if current is None:
                return head
            current = current.next

        # Delete N nodes
        if current is None:
            return head
        temp = current.next
        for _ in range(N):
            if temp is None:
                break
            next_node = temp.next
            temp = next_node

        # Connect previous node with the next node after deleting N nodes
        if current is None:
            return head
        current.next = temp
        current = temp

    return head


#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head




def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()
    m_n = stdin.readline().strip().split(" ")

    m = int(m_n[0])
    n = int(m_n[1])

    newHead = skipMdeleteN(head, m, n)
    printLinkedList(newHead)

    t -= 1
```

## Swap 2 Nodes of Linked List

You have been given a singly linked list of integers along with two integers, 'i,' and 'j.' Swap the nodes that are present at the 'i-th' and 'j-th' positions and return the new head to the list.

Note :
1. Remember, You need to swap the nodes, not only the data.
2. Indexing starts from 0.
3. No need to print the list, it has already been taken care.

```python
from sys import stdin

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None



def swapNodes(head, i, j):
    if head is None or head.next is None or i == j:
        return head

    # Initialize pointers
    prev_i, prev_j = None, None
    current_i, current_j = head, head
    index = 0

    # Traverse the list to find the i-th and j-th nodes and their previous nodes
    while current_i is not None and index != i:
        prev_i = current_i
        current_i = current_i.next
        index += 1

    index = 0
    while current_j is not None and index != j:
        prev_j = current_j
        current_j = current_j.next
        index += 1

    # If either i or j is out of range, or both nodes are the same, return the original head
    if current_i is None or current_j is None:
        return head

    # If i-th node is the head
    if prev_i is None:
        head = current_j
    else:
        prev_i.next = current_j

    # If j-th node is the head
    if prev_j is None:
        head = current_i
    else:
        prev_j.next = current_i

    # Swap next pointers
    temp = current_i.next
    current_i.next = current_j.next
    current_j.next = temp

    return head

#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head




def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()
    i_j = stdin.readline().strip().split(" ")

    i = int(i_j[0])
    j = int(i_j[1])

    newHead = swapNodes(head, i, j)
    printLinkedList(newHead)

    t -= 1
```

## kReverse

Example :
Given this linked list: 1 -> 2 -> 3 -> 4 -> 5

For k = 2, you should return: 2 -> 1 -> 4 -> 3 -> 5

For k = 3, you should return: 3 -> 2 -> 1 -> 5 -> 4
 Note :
No need to print the list, it has already been taken care. Only return the new head to the list.

## Bubble Sort Linked List - Iterative

Given a singly linked list of integers, sort it using 'Bubble Sort.'

Note :
No need to print the list, it has already been taken care. Only return the new head to the list.

```python

from sys import stdin

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None



def bubbleSort(head):
    if head is None or head.next is None:
        return head

    # Helper function to swap the data of two nodes
    def swap(node1, node2):
        temp = node1.data
        node1.data = node2.data
        node2.data = temp

    # Outer loop to iterate over the list
    while True:
        swapped = False
        current = head
        prev = None

        # Inner loop to compare and swap adjacent nodes
        while current and current.next:
            next_node = current.next

            # If current node's data is greater than next node's data, swap
            if current.data > next_node.data:
                swap(current, next_node)
                swapped = True

            prev = current
            current = current.next

        # If no swaps were made in this pass, the list is sorted
        if not swapped:
            break

    return head

def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head



def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
head = takeInput()
head = bubbleSort(head)
printLinkedList(head)
```
