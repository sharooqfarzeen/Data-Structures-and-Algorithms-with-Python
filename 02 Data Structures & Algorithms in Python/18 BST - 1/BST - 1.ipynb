## Elements between K1 and K2

Given a Binary Search Tree and two integers k1 and k2, find and print the elements which are in range k1 and k2 (both inclusive).

Print the elements in increasing order.

```python
class TreeNode(object):
    def __init__(self, x):
        self.data = x
        self.left = None
        self.right = None

class Store:
    def __init__(self):
        self.arr = []

s = Store()

def array_to_bst(array_nums):
    if not array_nums:
        return None
    mid_num = (len(array_nums)-1)//2
    node = TreeNode(array_nums[mid_num])
    node.left = array_to_bst(array_nums[:mid_num])
    node.right = array_to_bst(array_nums[mid_num+1:])
    return node

def elementsInRangeK1K2(root, k1, k2):
  if root == None:
    return
  if root.data < k1:
    elementsInRangeK1K2(root.right, k1, k2)
  elif root.data > k2:
    elementsInRangeK1K2(root.left, k1, k2)
  else:
    s.arr.append(root.data)
    elementsInRangeK1K2(root.left, k1, k2)
    elementsInRangeK1K2(root.right, k1, k2)

# Main
levelOrder = [int(i) for i in input().strip().split() if i != '-1']
levelOrder.sort()
root = array_to_bst(levelOrder)
k1, k2 = (int(i) for i in input().strip().split())
elementsInRangeK1K2(root, k1, k2)
s.arr.sort()
print(*s.arr)
```

## Construct BST

Given a sorted integer array A of size n, which contains all unique elements. You need to construct a balanced BST from this input array. Return the root of constructed BST.

Note: If array size is even, take first mid as root.

```python
class TreeNode:
    def __init__(self, x):
        self.data = x
        self.left = None
        self.right = None

def array_to_bst(array_nums):
    if not array_nums:
        return None
    mid_num = (len(array_nums)-1)//2
    node = TreeNode(array_nums[mid_num])
    node.left = array_to_bst(array_nums[:mid_num])
    node.right = array_to_bst(array_nums[mid_num+1:])
    return node

def preOrder(root):
  if root == None:
    return

  print(root.data, end = " ")
  preOrder(root.left)
  preOrder(root.right)

# Main
n = int(input())
if n > 0:
  levelOrder = [int(i) for i in input().split()]
  levelOrder.sort()
  root = array_to_bst(levelOrder)
  preOrder(root)
```
