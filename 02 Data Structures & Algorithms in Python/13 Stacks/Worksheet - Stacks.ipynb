## Stack Using Array

```python
class Stack:
  def __init__(self):
    self.__arr = []
    self.__top = -1
    self.__size = 0

  def push(self, data):
    self.__arr.append(data)
    self.__top += 1
    self.__size += 1

  def pop(self):
    if self.__size == 0:
      return -1

    element = self.__arr[self.__top]
    del self.__arr[self.__top]
    self.__top -= 1
    self.__size -= 1

    return element

  def top(self):
    if self.__size == 0:
      return -1

    return self.__arr[self.__top]

  def getSize(self):
    return self.__size

  def isEmpty(self):
    return self.__size == 0
```

```python
s = Stack()
```

```python
s.push(1)
s.push(2)
```
