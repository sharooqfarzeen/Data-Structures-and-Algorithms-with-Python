## Power of a Number

````python
def power(x, n):
    if n == 0:
        return 1
    if n == 1:
        return x
    if n % 2 == 0:
        small_power = power(x, n/2)
    else:
        small_power = power(x, n//2)
    return small_power * small_power

# Main
from sys import setrecursionlimit
setrecursionlimit(11000)
x, n=list(int(i) for i in input().strip().split(' '))
print(power(x, n))
````

## Array Equilibrium Index

````python
from sys import stdin

def arrayEquilibriumIndex(arr, n) :
    lsum = 0
    rsum = sum(arr[1:])
    if n == 0:
        return -1
    if rsum == 0:
        return 0
    for i in range(1,n):
        lsum += arr[i-1]
        rsum -= arr[i]
        if lsum == rsum:
            return i
    return -1


#Taking input using fast I/O method
def takeInput() :
    n = int(stdin.readline().strip())
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n


def printList(arr, n) :
    for i in range(n) :
        print(arr[i], end = " ")
    print()


#main
t = int(stdin.readline().strip())

while t > 0 :

    arr, n = takeInput()
    print(arrayEquilibriumIndex(arr, n))

    t-= 1
````

## Pair sum in Array

You have been given an integer array/list(ARR) and a number 'num'. Find and return the total number of pairs in the array/list which sum to 'num'.

Note:
Given array/list can contain duplicate elements.

````python
from sys import stdin

def pairSum(arr, n, num):
  arr.sort()
  count = 0

  if n <= 1:
    return 0

  for i in range(n):
    #current number in sorted array is already greater than pair sum
    if i > num:
      return count
    for j in range(i+1, n):
      sum = arr[i] + arr[j]
      if sum == num:
        count += 1
      elif sum > num:
        break
  return count

#taking input using fast I/O method
def takeInput() :
    n = int(stdin.readline().strip())
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n


def printList(arr, n) :
    for i in range(n) :
        print(arr[i], end = " ")
    print()


#main
t = int(stdin.readline().strip())

while t > 0 :

    arr, n = takeInput()
    num = int(stdin.readline().strip())
    print(pairSum(arr, n, num))

    t -= 1
````

## Triplet Sum

You have been given a random integer array/list(ARR) and a number X. Find and return the triplet(s) in the array/list which sum to X.

Note :
Given array/list can contain duplicate elements.

````python
#TA's solution
from sys import stdin

def tripletSum(arr, n, num):
    # Sort the array to simplify the search process
    arr.sort()

    # Initialize the variable to store the count of triplets
    count = 0

    # Iterate through the array elements up to the third-to-last element
    for i in range(n - 2):
        # Initialize two pointers, one pointing to the next element and another to the last element
        l = i + 1
        r = n - 1
        if arr[i] > num:
            return count

        # Use a two-pointer approach to find triplets
        while l < r:
            # Calculate the current sum of the triplet
            currsum = arr[i] + arr[l] + arr[r]

            # If the current sum is equal to the target num, we found a triplet
            if currsum == num:
                count += 1  # Increment the count of triplets
                tempr = r - 1

                # Handle duplicate elements in the array
                while tempr > l:
                    if arr[r] == arr[tempr]:
                        count += 1
                        tempr -= 1
                    else:
                        break

                l += 1

            # If the current sum is greater than the target, decrement the high pointer
            elif currsum > num:
                r -= 1
            # If the current sum is less than the target, increment the low pointer
            else:
                l += 1

    # Return the count of triplets
    return count


#taking input using fast I/O method
def takeInput() :
    n = int(stdin.readline().strip())
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n


def printList(arr, n) :
    for i in range(n) :
        print(arr[i], end = " ")
    print()


#main
t = int(stdin.readline().strip())

while t > 0 :

    arr, n = takeInput()
    num = int(stdin.readline().strip())
    print(tripletSum(arr, n, num))

    t -= 1
````
