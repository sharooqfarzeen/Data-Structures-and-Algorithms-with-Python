## Print Binary Tree

```python
class BinaryTreeNode:

  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None
```

```python
def printBT(root):
    if root == None:
      return

    print(root.data)
    printBT(root.left)
    printBT(root.right)
```

```python
_1_1 = BinaryTreeNode(1)
_2_1 = BinaryTreeNode(2)
_2_2 = BinaryTreeNode(3)
_3_1 = BinaryTreeNode(4)
_3_2 = BinaryTreeNode(100)
_3_3 = BinaryTreeNode(6)
_3_4 = BinaryTreeNode(7)
_4_1 = BinaryTreeNode(8)
_4_2 = BinaryTreeNode(9)
```

```python
printBT(root)
```

## Print Binary Tree Detailed

```python
def printBTD(root):
    if root == None:
      return

    print(root.data, ": ", sep = "", end = "")
    if root.left != None:
      print(root.left.data, end = "")
    if root.right != None:
      print(", ", root.right.data)
    printBTD(root.left)
    printBTD(root.right)
    print()
```

## Take Input

```python
def takeInput():
  rootdata = int(input())
  if rootdata == -1:
    return None
  root = BinaryTreeNode(rootdata)
  root.left = takeInput()
  root.right = takeInput()

  return root
```

```python
printBTD(root)
```

```python
printBTD(root)
```

## Number of nodes in a tree given the root

```python
def treeSize(root, count = 1):
  if root == None:
      return count - 1

  count = treeSize(root.left, count+1)
  count = treeSize(root.right, count+1)
  return count
```

```python
treeSize(root)
```

## Node with largest data

```python
def largestNode(root, largest = 0):
  if root == None:
    return largest
  largest = largestNode(root.left, largest)
  largest = largestNode(root.right, largest)
  if root.data > largest:
    return root.data
  return largest
```

```python
largestNode(root)
```

## Number of leaf nodes

```python
def leafCount(root):
  if root == None:
    return 0

  if root.left == None and root.right == None:
    return 1
  nl  = leafCount(root.left)
  nr = leafCount(root.right)
  return nl + nr
```

```python
leafCount(root)
```

## Print nodes at depth k

```python
def printDepth(root, k, depth = 0):
  if root == None:
    return

  printDepth(root.left, k, depth+1)
  printDepth(root.right, k, depth+1)

  if depth == k:
    print(root.data)
```

```python
printDepth(root, 2)
```

```python
printDepth(root, 0)
```
