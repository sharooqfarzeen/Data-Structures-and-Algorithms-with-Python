## Sum of all Nodes

Given a generic tree, find and return the sum of all nodes present in the given tree.

```python
import sys
import queue

class TreeNode :
    def __init__(self, data) :
        self.data = data
        self.children = list()

def inputLevelWise(li) :
    i = 0
    data = li[i]
    i += 1
    if data == -1 :
        return None
    root = TreeNode(data)
    q = queue.Queue()
    q.put(root)
    while (not q.empty()) :
        frontNode = q.get()
        noOfChildren = li[i]
        i += 1
        childrenArray = li[i : i+noOfChildren]
        for childData in childrenArray :
            childNode = TreeNode(childData)
            frontNode.children.append(childNode)
            q.put(childNode)
        i = i+noOfChildren
    return root

class Helper:
  def __init__(self):
    self.sum = 0

h = Helper()

def sumOfAllNodes(root):
  if  root == None:
    return h.sum

  h.sum += root.data

  for child in root.children:
    sumOfAllNodes(child)

  return h.sum

#main
sys.setrecursionlimit(10**6)
li = [int(elem) for elem in list(input().strip().split())]
root = inputLevelWise(li)
print(sumOfAllNodes(root))
```

## Node With Largest Data

Given a generic tree, find and return the node with maximum data. You need to return the node which is having maximum data.

Return null if tree is empty.

```python
import sys
sys.setrecursionlimit(3000)
from sys import stdin
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
    def __str__(self):
        return str(self.data)

class Helper:

  def __init__(self):
    self.max = float('-inf')

h = Helper()

def maxDataNode(root):
  if root == None:
    return

  if root.data > h.max:
    h.max = root.data

  for child in root.children:
    maxDataNode(child)

  return h.max

def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
arr = list(int(x) for x in stdin.readline().rstrip().rsplit())
tree = createLevelWiseTree(arr)
print(maxDataNode(tree))
```

## Height of Tree

Given a generic tree, find and return the height of given tree. The height of a tree is defined as the longest distance from root node to any of the leaf node. Assume the height of a tree with a single node is 1.

image.png

Height of above tree is 3.

```python
import sys
import queue

sys.setrecursionlimit(10**6)

class TreeNode:

  def __init__(self, data):

    self.data = data
    self.children = list()

def treeHeight(root):
  if root == None:
    return 0

  height = 0
  temp = 0

  for child in root.children:
    temp = treeHeight(child)
    if temp > height:
        height = temp

  return height + 1



def takeInputLW():
  #taking string input, converting to integer list and reversing the list...
  #for pop to work without keeping an index variable
  li = [int(i) for i in input().strip().split()][::-1]
  q = queue.Queue()

  rootData = li.pop()

  if rootData != -1:
    root = TreeNode(rootData)
    q.put(root)

    while not q.empty():
      currNode = q.get()
      nChild = li.pop()
      while nChild > 0:
        childData = li.pop()
        childNode = TreeNode(childData)
        currNode.children.append(childNode)
        q.put(childNode)
        nChild -= 1

  return root

root = takeInputLW()
print(treeHeight(root))
```

## Print Levelwise

Given a generic tree, print the input tree in level wise order.

For printing a node with data N, you need to follow the exact format -

N:x1,x2,x3,...,xn
where, N is data of any node present in the generic tree. x1, x2, x3, ...., xn are the children of node N. Note that there is no space in between.
You need to print all nodes in the level order form in different lines.

```python
import queue

def printLevelWiseTree():
  li = [int(i) for i in input().strip().split()][::-1]
  q = queue.Queue()

  rootData = li.pop()
  q.put(rootData)

  while not q.empty():
    currNode = q.get()
    print(currNode, end = ":")
    nChild = li.pop()
    while nChild > 0:
      childData = li.pop()
      print(childData, end = "")
      if nChild != 1:
        print(",", end = "")
      q.put(childData)
      nChild -= 1
    print()

printLevelWiseTree()
```

## Contains x

Given a generic tree and an integer x, check if x is present in the given tree or not. Return true if x is present, return false otherwise.

```python
from sys import stdin,setrecursionlimit
import queue
setrecursionlimit(10**6)
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

class Helper:
  def __init__(self):
    self.found = False

h = Helper()

def containsX(tree, x):

  if tree.data == x:
    h.found = True

  if not h.found:
    for child in tree.children:
      containsX(child, x)

def takeInputLW():
  #taking string input, converting to integer list and reversing the list...
  #for pop to work without keeping an index variable
  li = [int(i) for i in input().strip().split()][::-1]
  q = queue.Queue()

  rootData = li.pop()

  if rootData != -1:
    root = TreeNode(rootData)
    q.put(root)

    while not q.empty():
      currNode = q.get()
      nChild = li.pop()
      while nChild > 0:
        childData = li.pop()
        childNode = TreeNode(childData)
        currNode.children.append(childNode)
        q.put(childNode)
        nChild -= 1

  return root

# Main
tree = takeInputLW()
x = int(input())
containsX(tree, x)
print("true") if h.found else print("false")
```

## Node with maximum child sum

Given a generic tree, find and return the node for which sum of its data and data of all its child nodes is maximum. In the sum, data of the node and data of its immediate child nodes has to be taken.

```python
from sys import stdin,setrecursionlimit
import queue
setrecursionlimit(10**6)
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

class Helper:
  def __init__(self):
    self.maxSum = float('-inf')
    self.maxSumNode = -1

h = Helper()

def maxSum(tree):
  nodeSum = tree.data
  for child in tree.children:
    nodeSum += child.data
  if nodeSum > h.maxSum:
    h.maxSum = nodeSum
    h.maxSumNode = tree.data
  for child in tree.children:
    maxSum(child)

def takeInputLW():
  #taking string input, converting to integer list and reversing the list...
  #for pop to work without keeping an index variable
  li = [int(i) for i in input().strip().split()][::-1]
  q = queue.Queue()

  rootData = li.pop()

  if rootData != -1:
    root = TreeNode(rootData)
    q.put(root)

    while not q.empty():
      currNode = q.get()
      nChild = li.pop()
      while nChild > 0:
        childData = li.pop()
        childNode = TreeNode(childData)
        currNode.children.append(childNode)
        q.put(childNode)
        nChild -= 1

  return root

# Main
tree = takeInputLW()
maxSum(tree)
print(h.maxSumNode)
```

## Structurally Identical

Given two generic trees, return true if they are structurally identical. Otherwise return false.

Structural Identical
If the two given trees are made of nodes with the same values and the nodes are arranged in the same way, then the trees are called identical.

```python
from sys import stdin,setrecursionlimit
import queue
setrecursionlimit(10**6)
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

class Helper:
  def __init__(self):
    self.identical = True

h = Helper()

def identical(tree1, tree2):

  if tree1.data != tree2.data:
    h.identical = False

  if h.identical:
    i1 = len(tree1.children)
    i2 = len(tree2.children)
    if i1 == i2:
      for i in range(i1):
        identical(tree1.children[i], tree2.children[i])
    else:
      h.identical = False

def takeInputLW():
  #taking string input, converting to integer list and reversing the list...
  #for pop to work without keeping an index variable
  li = [int(i) for i in input().strip().split()][::-1]
  q = queue.Queue()

  rootData = li.pop()

  if rootData != -1:
    root = TreeNode(rootData)
    q.put(root)

    while not q.empty():
      currNode = q.get()
      nChild = li.pop()
      while nChild > 0:
        childData = li.pop()
        childNode = TreeNode(childData)
        currNode.children.append(childNode)
        q.put(childNode)
        nChild -= 1

  return root

# Main
tree1 = takeInputLW()
tree2 = takeInputLW()
identical(tree1, tree2)
print("true") if h.identical else print("false")
```

## Next Larger

Given a generic tree and an integer n. Find and return the node with next larger element in the tree i.e. find a node with value just greater than n.

Note: Return NULL if no node is present with the value greater than n.

```python
from sys import stdin,setrecursionlimit
import queue
setrecursionlimit(10**6)
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

class Helper:
  def __init__(self):
    self.nextLarger = float('inf')

h = Helper()

def nextLarger(tree, n):
  if tree.data > n and tree.data < h.nextLarger:
    h.nextLarger = tree.data

  for child in tree.children:
    nextLarger(child, n)

def takeInputLW():
  #taking string input, converting to integer list and reversing the list...
  #for pop to work without keeping an index variable
  li = [int(i) for i in input().strip().split()][::-1]
  q = queue.Queue()

  rootData = li.pop()

  if rootData != -1:
    root = TreeNode(rootData)
    q.put(root)

    while not q.empty():
      currNode = q.get()
      nChild = li.pop()
      while nChild > 0:
        childData = li.pop()
        childNode = TreeNode(childData)
        currNode.children.append(childNode)
        q.put(childNode)
        nChild -= 1

  return root

# Main
tree = takeInputLW()
n = int(input())
nextLarger(tree, n)
if h.nextLarger != float('inf'):
  print(h.nextLarger)
```

## Replace with depth

You are given a generic tree. You have to replace each node with its depth value. You just have to update the data of each node, there is no need to return or print anything.

Sample Input 1:
10 3 20 30 40 2 40 50 0 0 0 0
Sample Output 1:
0
1 1 1
2 2

```python
from sys import stdin,setrecursionlimit
import queue
setrecursionlimit(10**6)
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

def replacewithDepth(tree, depth = 0):
  tree.data = depth

  for child in tree.children:
    replacewithDepth(child, depth + 1)


def takeInputLW():
  #taking string input, converting to integer list and reversing the list...
  #for pop to work without keeping an index variable
  li = [int(i) for i in input().strip().split()][::-1]
  q = queue.Queue()

  rootData = li.pop()

  if rootData != -1:
    root = TreeNode(rootData)
    q.put(root)

    while not q.empty():
      currNode = q.get()
      nChild = li.pop()
      while nChild > 0:
        childData = li.pop()
        childNode = TreeNode(childData)
        currNode.children.append(childNode)
        q.put(childNode)
        nChild -= 1

  return root

def printLevelAtNewLine(tree):
    q = [tree]
    newq = []
    while q:
        parent = q.pop(0)
        print(parent.data, end=' ')
        for child in parent.children:
            newq.append(child)
        if len(q)==0:
            q = newq
            newq = []
            print()  # Move to next Line

# Main
tree = takeInputLW()
replacewithDepth(tree)
printLevelAtNewLine(tree)
```
