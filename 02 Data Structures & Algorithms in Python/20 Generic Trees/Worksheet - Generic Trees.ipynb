```python
class GenericTreeNode:
  def __init__(self, data):
    self.data = data
    self.children = list()
```

```python
n1 = GenericTreeNode(1)
n2 = GenericTreeNode(2)
n3 = GenericTreeNode(3)
n4 = GenericTreeNode(4)
n5 = GenericTreeNode(5)
n6 = GenericTreeNode(6)
n7 = GenericTreeNode(7)

n1.children.extend([n2, n3, n4, n5])
n3.children.extend([n6, n7])
```

```python
def printGT(root):
  if root == None:
    return

  print(root.data, end = ":")

  if root.children:
    [print(child.data, end = " ") for child in root.children]
  print()

  [printGT(node) for node in root.children]
```

```python
printGT(n1)
```

```python
def takeInput():
  print("Enter root")
  rootData = int(input())
  if rootData == -1:
    return

  root = GenericTreeNode(rootData)
  print("Enter number of children for", rootData)
  childrenCount = int(input())

  for i in range(childrenCount):
    child = takeInput()
    root.children.append(child)

  return root
```

```python
root=takeInput()
printGT(root)
```
