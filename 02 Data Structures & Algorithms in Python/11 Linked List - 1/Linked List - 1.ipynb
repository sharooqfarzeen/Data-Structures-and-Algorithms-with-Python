## Delete Node Recursively

 Note :
Assume that the Indexing for the linked list always starts from 0.

No need to print the list, it has already been taken care. Only return the new head to the list.

```python
from sys import stdin, setrecursionlimit
setrecursionlimit(10 ** 6)

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None


def deleteNodeRec(head, pos, i = 0, prev = None):
  if i == pos:
    if pos == 0:
      head = head.next
      return head
    else:
      prev.next = head.next
      return
  if head.next == None:
    return
  prev = head
  deleteNodeRec(head.next, pos, i+1, prev)
  return head

def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head




def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()
    pos = int(stdin.readline().rstrip())

    newHead = deleteNodeRec(head, pos)
    printLinkedList(newHead)

    t -= 1
```

## Eliminate Duplicates

You have been given a singly linked list of integers where the elements are sorted in ascending order. Write a function that removes the consecutive duplicate values such that the given list only contains unique elements and returns the head to the updated list.

```python
from sys import stdin

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None

def findLength(head) :
    count = 0
    if head.next is None:
      return 0
    else:
      count += 1
    while head.next is not None:
      count += 1
      head = head.next
    return count

def removeDuplicates(head) :
    if head == None:
      return head
    if head.next == None:
      return head

    length = findLength(head)

    curr = head.next
    prev = head
#1 1 1 2 3 3
    position = 1
    while position < length:
      if curr.data == prev.data:
        # print("Del",curr.data,"at",position)
        prev.next = curr.next
        curr = prev.next
        position += 1
        continue
      prev = prev.next
      curr = curr.next
      position += 1

    return head

#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = [int(i) for i in input().split()]

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head


#to print the linked list
def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(input())

while t > 0 :
    head = takeInput()

    head = removeDuplicates(head)
    printLinkedList(head)

    t -= 1
```

To explain it further, you need to start printing the data from the tail and move towards the head of the list, printing the head data at the end.

Note :
You can’t change any of the pointers in the linked list, just print the linked list in the reverse order.

```python

from sys import stdin, setrecursionlimit
setrecursionlimit(10 ** 6)

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None




def printReverse(head):
  if head == None:
    return
  if head.next == None:
    print(head.data, end = " ")
    return
  printReverse(head.next)
  print(head.data, end = " ")
  return

#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head


#to print the linked list
def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()
    printReverse(head)
    print()

    t -= 1
```

## Palindrome Linked List

You have been given a head to a singly linked list of integers. Write a function check to whether the list given is a 'Palindrome' or not.

```python

from sys import stdin, setrecursionlimit
setrecursionlimit(10**5)

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None



def isPalindrome(head):
  if head == None:
    return True
  if head.next == None:
    return True

  li = []

  while head is not None:
    li.append(head.data)
    head = head.next

  return True if li == li[::-1] else return False


#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = list(map(int, stdin.readline().rstrip().split(" ")))

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head


#to print the linked list
def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    head = takeInput()

    if isPalindrome(head) :
        print('true')
    else :
        print('false')

    t -= 1
```
