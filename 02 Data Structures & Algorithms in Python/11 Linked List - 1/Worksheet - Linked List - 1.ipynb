Space separated integer input as linked list. End of list denoted by -1.

Head --> Data1, Next1 --> Data2, Next2 --> ....... --> Datan, None <-- Tail

```python
class Node:
  def __init__(self, number):
    self.number = number
    self.next = None

def takeInput():

  numbers = [int(i) for i in input().split()]
  head = None
  tail = None

  for data in numbers:
    if data == -1:
      break
    newNode = Node(data)
    if head is None:
      head = newNode
      tail = newNode
    else:
      tail.next = newNode
      tail = newNode
  return head

def printLL(head):
  while head is not None:
    print(head.number, "-->1 2", end = " ")
    head = head.next

head = takeInput()
printLL(head)
```
