## Fibonacci Recursive - Using Dynamic Programming

```python
def fib(n):
  if n == 1 or n == 0:
    return n
  elif memo[n] != -1:
    return memo[n]
  else:
    memo[n] = fib(n-1) + fib(n-2)
    return memo[n]

#main
n = int(input())

#array for memoization
memo = [-1] * (n+1)

print("Nth Fibonacci Number is:", fib(n))
```
