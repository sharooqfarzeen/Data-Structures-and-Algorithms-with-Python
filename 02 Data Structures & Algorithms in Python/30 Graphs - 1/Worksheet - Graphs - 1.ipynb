## DFS

```python
class Graph:

  def __init__(self, n):
    # Dictionary with nodes as keys, each key gets its own list
    self.graph = {i: [] for i in range(1, n+1)}
    self.visited = [False] * (n+1)

  def add_edge(self, u, v):
    #edge from u to v
    self.graph[u].append(v)
    #edge from v to u
    self.graph[v].append(u)

  def print(self):
    print(self.graph)

  def dfs(self, s):
    print("At" + str(s))
    self.visited[s] = True
    for node in self.graph[s]:
      if self.visited[node] == False:
        self.dfs(node)



g = Graph(4)
g.add_edge(1,2)
g.add_edge(1,3)
g.add_edge(2,3)
g.add_edge(2,4)
g.print()
```

```python
g.dfs(1)
```
