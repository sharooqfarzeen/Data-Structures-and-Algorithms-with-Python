## Is Tree Balanced?

Tree is balanced when difference of heights of sub-trees of all nodes are less than or equal to 1

```python
def height(root):
  if root == None:
    return 0
  return max(height(root.left), height(root.right)) + 1
```

```python
def isBalanced(root):
  if root == None:
    return True

  leftHeight = height(root.left)
  rightHeight = height(root.right)

  if abs(leftHeight - rightHeight) > 1:
    return False

  isLeftBalanced = isBalanced(root.left)
  isRightBalanced = isBalanced(root.right)

  if isLeftBalanced and isRightBalanced:
    return True
  else:
    return False
```

## Taking Input - Level wise

```python
import queue
def levelWiseInput():
  print("Enter root:", end = " ")
  rootData = int(input())
  print()
  if rootData != -1:
    #creating root node
    root = BinaryTreeNode(rootData)
  else:
    #empty tree
    return None
  #queue to push non-empty nodes
  q = queue.Queue()
  q.put(root)
  while not q.empty():
    curr_node = q.get()
    #receiving and handling left child
    print("Enter left child of ", curr_node.data, ": ", sep = "", end = " ")
    leftChildData = int(input())
    print()
    if leftChildData != -1:
      leftChild = BinaryTreeNode(leftChildData)
      curr_node.left = leftChild
      q.put(leftChild)
    #receiving and handling right child
    print("Enter right child of ", curr_node.data, ": ", sep = "", end = " ")
    rightChildData = int(input())
    print()
    if rightChildData != -1:
      rightChild = BinaryTreeNode(rightChildData)
      curr_node.right = rightChild
      q.put(rightChild)
  return root
```

```python
root = levelWiseInput()
printBTD(root)
```

```python
def printLevelWise(root):
  #empty tree
  if root == None:
    return
  q = queue.Queue()
  q.put(root)
  while not q.empty():
    currNode = q.get()
    print(currNode.data, end = ":")
    #left child
    if currNode.left != None:
      print("L:",currNode.left.data, sep = "", end = ",")
      q.put(currNode.left)
    else:
      print("L:", -1, sep = "", end = ",")
    #right child
    if currNode.right != None:
      print("R:", currNode.right.data, sep = "")
      q.put(currNode.right)
    else:
      print("R:", -1, sep = "")
```
