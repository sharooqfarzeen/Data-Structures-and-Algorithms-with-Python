## Maximum Frequency

You are given an array of integers that contain numbers in random order. Write a program to find and return the number which occurs the maximum times in the given input.

If two or more elements are having the maximum frequency, return the element which occurs in the array first.

```python
from collections import OrderedDict
#taking inputs
n = int(input())

#empty dict
d = OrderedDict()

for i in input().strip().split():
    if int(i) in d:
        d[int(i)] += 1
    else:
        d[int(i)] = 1

keyList = list(d.keys())
valueList = list(d.values())

print(keyList[valueList.index(max(valueList))])
```

## Pairs with difference k

You are given with an array of integers and an integer K. You have to find and print the count of all such pairs which have difference K.

Note: Take absolute difference between the elements of the array.

```python
def printPairDiffK(l, k):
  d = {}
  count = 0
  length = len(l)

  for i in range(length):
    for j in l[i+1:]:
      if abs(l[i] - j) == k:
        count += 1
  return count
# Main
n=int(input())
l=list(int(i) for i in input().strip().split(' '))
k=int(input())
print(printPairDiffK(l, k))
```

```python
def printPairDiffK(l, k):
  d = {}
  count = 0

  for num in l:
    greater = num + k
    lesser = num - k
    count += d.get(greater, 0)
    if k != 0:
      count += d.get(lesser, 0)

    if d.get(num):
      d[num] += 1
    else:
      d[num] = 1

  return count

# Main
n=int(input())
l=list(int(i) for i in input().strip().split(' '))
k=int(input())
print(printPairDiffK(l, k))
```

## Longest Consecutive Sequence

You are given an unsorted array/list 'ARR' of 'N' integers. Your task is to return the length of the longest consecutive sequence.

The consecutive sequence is in the form ['NUM', 'NUM' + 1, 'NUM' + 2, ..., 'NUM' + L] where 'NUM' is the starting integer of the sequence and 'L' + 1 is the length of the sequence.

Note:

If there are any duplicates in the given array we will count only one of them in the consecutive sequence.
For example-
For the given 'ARR' [9,5,4,9,10,10,6].

Output = 3
The longest consecutive sequence is [4,5,6].
Follow Up:
Can you solve this in O(N) time and O(N) space complexity?

[4,5,6,9,9,10,10]

```python
def lengthOfLongestConsecutiveSequence(arr, n):
    arr.sort()
    maxCount, count, i = 0, 0, 0
    while i < n:
      while (i < n-1) and (arr[i+1] in (arr[i], arr[i]+1)):
        if arr[i+1] != arr[i]:
          count += 1
        i += 1
      if count > maxCount:
        maxCount = count
      count = 0
      i += 1
    return maxCount + 1
```

## Extract Unique Characters

Given a string S, you need to remove all the duplicates. That means, the output string should contain each character only once. The respective order of characters should remain same, as in the input string.

## Longest subarray sum zero

Given an array consisting of positive and negative integers, find the length of the longest subarray whose sum is zero.

```python
def subsetSum(arr):
    # Dictionary to store the running sum and its index
    sum_map = {}
    max_length = 0
    current_sum = 0

    # Traverse the array
    for i in range(len(arr)):
        # Add the current element to the running sum
        current_sum += arr[i]

        # If the current sum is zero, update the max length
        if current_sum == 0:
            max_length = i + 1

        # If the current sum is already in the dictionary,
        # update the max length with the difference between
        # the current index and the index of the same sum
        elif current_sum in sum_map:
            max_length = max(max_length, i - sum_map[current_sum])

        # If the current sum is not in the dictionary,
        # store it along with its index
        else:
            sum_map[current_sum] = i

    return max_length

# Main
n=int(input())
l=list(int(i) for i in input().strip().split(' '))
print(subsetSum(l))
```

## Print Intersection

You have been given two integer arrays/lists (ARR1 and ARR2) of size N and M, respectively. You need to print their intersection; An intersection for this problem can be defined when both the arrays/lists contain a particular value or to put it in other words, when there is a common value that exists in both the arrays/lists.

Note :
Input arrays/lists can contain duplicate elements.

The intersection elements printed would be in the order they appear in the second array/list (ARR2)

```python
from sys import stdin

def printIntersection(arr1, n1, arr2, n2):
  count1 = {}
  for ele in arr1:
    count1[ele] = count1.get(ele, 0) + 1

  for ele in arr2:
    if (ele in count1) and (count1[ele] > 0):
      print(ele)
      count1[ele] -= 1

#Taking input using fast I/O method
def takeInput() :
    n = int(stdin.readline().strip())
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n

#Main Code
arr1, n1 = takeInput()
arr2, n2 = takeInput()
printIntersection(arr1, n1, arr2, n2)
```
