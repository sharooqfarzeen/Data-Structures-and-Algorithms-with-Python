## Row Wise Sum

```python
def rowWiseSum(mat, nRows, mCols):
    for i in range(nRows):
        sum = 0
        for j in range(mCols):
            sum += mat[i][j]
        print(sum, end = ' ')
```

## Largest Row or Column

```python
def findLargest(arr, nRows, mCols):
    r_c = "row"
    rc_no = 0
    max_sum = -2147483648
    
    for i in range(nRows):
        sum = 0
        for j in range(mCols):
            sum += arr[i][j]
        if sum > max_sum:
            rc_no = i
            max_sum = sum
    
    for j in range(mCols):
        sum = 0
        for i in range(nRows):
            sum += arr[i][j]
        if sum > max_sum:
            r_c = "column"
            rc_no = j
            max_sum = sum
    
    print(r_c, rc_no, max_sum)
```

## Wave Print

```python
def wavePrint(mat, nRows, mCols):
    j = 0
    while j < mCols:
        if j%2 == 0:
            i = 0
            while i < nRows:
                print(mat[i][j], end = ' ')
                i += 1
        else:
            i = nRows-1
            while i >= 0:
                print(mat[i][j], end = ' ')
                i -= 1
    
wavePrint([[1,2,3,4],
           [5,6,7,8]], 2, 4)
```
