## Code : Square Pattern

No: of rows: n
  No: columns in ith row: n
  What to print: n

```python
n =  int(input())
i = 0
j = 0

while i < n:
    while j < n:
        print(n, end = '')
        j += 1
    j = 0
    i += 1
    print()
```

### Code : Triangular Star Pattern

*
**
***
****

no of rows = n
no of columns in ith row = i + 1
what to print? *

```python
n = int(input())

for i in range(n):
    for j in range(i+1):
        print("*", end = '')
    print()
```

### Code : Reverse Number Pattern

1
21
321
4321

number of rows = n
number of columns in ith row = i + 1
what to print? i-j+1

```python
n = int(input())

for i in range(n):
    t = i
    for j in range(i+1):
        print(t+1, end = '')
        t -= 1
    print()
```

```python
n = int(input())

for i in range(n):
    for j in range(i+1):
        print(i-j+1, end = '')
    print()
```

### Code : Character Pattern

A
BC
CDE
DEFG

number of rows = n
number of columns in ith row = i+1
what to print: first_char + i

```python
n = int(input())

first_char = ord('A')

for i in range(n):
    t = i
    for j in range(i+1):
        print(chr(first_char+t), end = '')
        t += 1
    print()
```

### Code : Interesting Alphabets

Pattern for N = 5
E
DE
CDE
BCDE
ABCDE

number of rows = n
number of columns in ith row = i+1
what to print? first_char-i

```python
n = int(input())

#ASCII of A
A = ord('A')
#ASCII of first character in pattern
first_char = A + n - 1

for i in range(n):
    t = i
    for j in range(i+1):
        print(chr(first_char-t), end = '')
        t -= 1
    print()
```

```python
#taking input
n = int(input())

print(1)

for i in range(1, n):
    print(i, end = '')
    j = i
    for j in range(i-1):
        print(0, end = '')
    print(i)
```
